```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
dados = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Elo Science - Semana Data Science/sds_fifa_atualizado.csv')
dados.head()
```

```
dados.plot()
```

```
dados.hist()
```

```
# Informações de correlação
dados.corr()
```

```
colunas = dados.columns
```

Mapa de Calor

```
plt.figure(figsize=(28, 17))
```

```
sns.heatmap(dados[colunas].corr(), annot=True)
```

Cluster map

```
sns.clustermap(dados[colunas].corr(), annot=False)
```

Box Splot

```
plt.figure(figsize=(10, 5))
pl = sns.boxplot(x=dados.Time[:40], y=dados.Media, palette='hls');
pl.set_title(label='Média por Time');
plt.xlabel('Time')
plt.ylabel('Média')
plt.yticks(rotation=80)
plt.ylim(85, 95)
```

Join plot

```
sns.jointplot(dados.Idade, dados.Media)
```

```
sns.jointplot(dados.Media, dados.Salario)
```

Cat Plot

```
sns.catplot('Idade', 'Media', data=dados)
```

```
sns.catplot('Idade', 'Media', data=dados, kind='box').fig.set_size_inches(11, 7)
```
